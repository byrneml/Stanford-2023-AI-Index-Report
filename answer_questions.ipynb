```
from langchain.document_loaders import UnstructuredPDFLoader, OnlinePDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter


loader = UnstructuredPDFLoader('./HAI_REPORT.pdf')
data = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=0)
texts = text_splitter.split_documents(data)
```

```
import os
from dotenv import load_dotenv
from langchain.vectorstores import Pinecone
from langchain.embeddings.openai import OpenAIEmbeddings
import pinecone

load_dotenv()

pinecone.init(
    api_key=os.getenv('PINECONE_KEY'),
    environment='us-east4-gcp'
)

index_name = 'stanford-report'
```

## Ask a Question

```
query = 'What is the most controversial opinion in this paper?'
```

```
from langchain.llms import OpenAI
from langchain.chains.question_answering import load_qa_chain

embeddings = OpenAIEmbeddings(openai_api_key=os.getenv("OPENAI_KEY"))

docsearch = Pinecone.from_existing_index(index_name=index_name, embedding=embeddings)

llm = OpenAI(temperature=0, openai_api_key=os.getenv("OPENAI_KEY"))
chain = load_qa_chain(llm, chain_type="stuff")

docs = docsearch.similarity_search(query, include_metadata=True)

res = chain.run(input_documents=docs, question=query)
print(res)
```
